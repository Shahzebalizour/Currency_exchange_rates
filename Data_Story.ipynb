```python
# Importing Libraries 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""We have dataset about currencies of different contries with exchange range of euro, It means Euro to Other currencies 
    So now we will compare about euro to usd so we will check the every possible changes of usd then when it was above from 
    euro and when it decreases and what are the reasons about usd got down from euro, so we will investigate about these
    things"""

# Import dataset
dataset = pd.read_csv("euro-daily-hist_1999_2022.csv")
dataset
```

```python
# Check the summary of our dataset
dataset.info()
```

```python
# Rename the column names of usd and time
dataset.rename(columns = {"Period\\Unit:" : "Time", "[US dollar ]" : "US_dollar"}, inplace = True)
```

```python
# Now convert Time column to datetime
dataset["Time"] = pd.to_datetime(dataset["Time"])
```

```python
# Sort the values by time
dataset.sort_values("Time", inplace = True)
```

```python
dataset.head()
```

```python
# Now we will separate those column on which we have to perform the analysis. So those columns are Time and USD
euro_dollar = dataset[["Time", "US_dollar"]].copy()
euro_dollar["US_dollar"].value_counts()
# We can see the messy values of - it may be holidays or something so we don't want this
```

```python
# We only need those rows which are not equal to -
euro_dollar = euro_dollar[euro_dollar["US_dollar"] != "-"]
euro_dollar["US_dollar"] = euro_dollar["US_dollar"].astype(float)
```

```python
euro_dollar.info()
```

# Rolling Mean

```python
plt.plot(euro_dollar["Time"], euro_dollar["US_dollar"])

plt.show()

print("As we can see that there are too many wiggles in our graph due to daily variation, so our graph is representable not\
not giving us good insights however we can see that the rates goes up and down, up and down again day to day. The rate only \
show clear upword or downward when we will see the data by months or yearly gaps, if we want to show only long term trends \
then we will use Rolling Mean or we can call it Moving Average")
```

```python
# Understanding the rolling mean concept
daily_values = pd.DataFrame()
daily_values["Values"] = pd.Series(range(1,20,2))
daily_values
```

```python
daily_values["rolling_mean_2"] = daily_values["Values"].rolling(2).mean()
daily_values
```

```python
plt.figure(figsize = (10, 8))

plt.subplot(3,2,1)
plt.plot(euro_dollar["Time"], euro_dollar["US_dollar"])
plt.title("Original Values", weight = 'bold')

for i, rolling_mean in zip([2,3,4,5,6],
                           [7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_dollar["Time"], euro_dollar["US_dollar"].rolling(rolling_mean).mean())
    plt.title("Rolling Window:" + str(rolling_mean), weight = 'bold')

plt.tight_layout()
plt.show()
```

# Coming up with an Idea
We show how the Euro-dollar has changed during corona pandemic duration. We can also show the 2020 data and 2016-2019 data for comparison line based. We will use line plot

We will show how the euro-dollar rate has changed during 2007-2008 financial crisis. We can also show the data of 2006-2009 for comparison 

We will show comparatively how the euro-dollar rate has changed under the last three US Presidents (George W Bush (2001-2009),
Barack Obama (2009-2017), and donald trump (2017-2021)

```python
euro_dollar["rolling_mean"] = euro_dollar["US_dollar"].rolling(30).mean()
```

```python
crisis_7_8 = euro_dollar.copy()[(euro_dollar["Time"].dt.year >= 2007) & (euro_dollar["Time"].dt.year <= 2008)]
crisis_6_9 = euro_dollar.copy()[(euro_dollar.Time.dt.year >= 2006) & (euro_dollar.Time.dt.year <= 2009)]
```

```python
# Create a graph for Financial Crisis
plt.style.use("fivethirtyeight")

# Adding the plot
fig, ax = plt.subplots(figsize = (9,3))
ax.plot(crisis_6_9["Time"], crisis_6_9["rolling_mean"], linewidth = 1, color = "#eaa9fc")


# Highlighting Financial Crisis
ax.plot(crisis_7_8["Time"], crisis_7_8["rolling_mean"], color = "#554d7a", linewidth = 3)

# Removing Xtixks and Yticks
ax.set_xticklabels([])
ax.set_yticklabels([])

# Adding x-axis labels
x = 0.02
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, -0.08, year, alpha = 0.5, fontsize = 11, transform = plt.gca().transAxes)
    x += 0.22888
    

# Adding Y-ticks
y = 0.07
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(-.04, y, rate, alpha = 0.5, fontsize = 11, transform = plt.gca().transAxes)
    y += 0.2333
    
# Adding title and subtitle
ax.text(-.05, 1.2, "There was 1.59 peak of Euro to USD while 2007-2008's Financial Crisis",
       weight = 'bold', transform = plt.gca().transAxes)
ax.text(-.05, 1.1, "Euro-USD exchange rates from 2006-2010", 
       size = 12, transform = plt.gca().transAxes)

# Adding text signature
ax.text(-.05, -0.25, "@TECHMA ZONE" + ' '*80 + "Source: European Central Bank", 
       size = 12, color = "#666466", backgroundcolor = "#99a3a2", transform = plt.gca().transAxes)

# Highlighting Those months which have highest peak of US Dollar
ax.axvspan(xmin = pd.to_datetime("2008-04-1"), xmax = pd.to_datetime("2008-9-1"), ymin = 0.09,
          alpha = 0.3, color = 'gray')

plt.show()
```

#### Let's Observe Corona Virus Pandemic Duration

```python
corona_20 = euro_dollar[(euro_dollar["Time"].dt.year >= 2020) & (euro_dollar["Time"].dt.year <= 2020)]
corona_comp = euro_dollar[(euro_dollar["Time"].dt.year >= 2016) & (euro_dollar["Time"].dt.year <= 2019)]
```

```python
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots(figsize = (10,4))

# Create a graph
ax.plot(corona_comp["Time"], corona_comp["rolling_mean"], linewidth = 1.5, color = "#a2e8cb")

# Adding Corona Virus Duration
ax.plot(corona_20["Time"], corona_20["rolling_mean"], linewidth = 3, color = "#225751")

# Removing labels
ax.set_xticklabels([])
ax.set_yticklabels([])

# Add X-axis Values
x = 0.02
for year in ['2016', '2017', '2018', '2019', '2020', '2021']:
    ax.text(x, -0.08, year, fontsize = 12, alpha = 0.5, transform = plt.gca().transAxes)
    x += 0.183

# Adding Y-axis Values
y = 0.02
for rate in ['1.05', '1.10', '1.15', '1.20']:
    ax.text(-.05, y, rate, alpha = 0.5, fontsize = 11, transform = plt.gca().transAxes)
    y += 0.248

# Adding title and subtitle
ax.text(-.05, 1.09, "The peak of Euro-USD was 1.22 during Covid 2020", weight = 'bold', transform = plt.gca().transAxes)
ax.text(-.05, 1.03, "Exchange rates of Euro-USD between 2016 and 2020", size = 12, transform = plt.gca().transAxes)

# Adding Signature 
ax.text(-.05, -0.21, "©TECHMA ZONE" + ' '*100 + "Source: European Central Bank", size = 12,
        color = "#f0f7f7", backgroundcolor = "#223634", transform = plt.gca().transAxes)

plt.show()
```

```python
# Now separate the data of those three presidents 
bush_obama_trump = euro_dollar[(euro_dollar["Time"].dt.year >= 2001) & (euro_dollar["Time"].dt.year < 2021)]

# Separate the data of bush 
bush = bush_obama_trump[(bush_obama_trump.Time.dt.year >= 2001) & (bush_obama_trump.Time.dt.year < 2009)]
obama = bush_obama_trump[(bush_obama_trump.Time.dt.year >= 2009) & (bush_obama_trump.Time.dt.year < 2017)]
trump = bush_obama_trump[(bush_obama_trump.Time.dt.year >= 2017) & (bush_obama_trump.Time.dt.year < 2021)]
```

```python
plt.style.use("fivethirtyeight")

plt.figure(figsize = (12, 6))

# Pattern 1
ax1 = plt.subplot(3,3,1)
ax2 = plt.subplot(3,3,2)
ax3 = plt.subplot(3,3,3)

# Pattern 2
ax4 = plt.subplot(3,1,2)

# Change the labels of Y and X axis
for ax in [ax1, ax2, ax3, ax4]:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6 $'], alpha = 0.4)

# Ax1 for Bush
ax1.plot(bush["Time"], bush["rolling_mean"], color = "#FFEA00", linewidth = 3.5)
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha = 0.4)
ax1.text(0.112, 2.45, "BUSH", weight = 'bold', fontsize = 15, color = "#FFEA00", transform = plt.gca().transAxes)
ax1.text(0.080, 2.30, '(2001-2009)', weight = 'bold', alpha = 0.3, transform = plt.gca().transAxes)

# Ax2 For Obama
ax2.plot(obama["Time"], obama["rolling_mean"], linewidth = 3.5, color = "#06BFF1")
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '', '2015', '', '2017'], alpha = 0.4)
ax2.text(0.458, 2.45, "OBAMA", color = "#06BFF1", fontsize = 15, weight = 'bold', transform = plt.gca().transAxes)
ax2.text(0.438, 2.30, "(2009-2017)", weight = 'bold', alpha = 0.3, transform = plt.gca().transAxes)

# Ax3 For Trump
ax3.plot(trump["Time"], trump["rolling_mean"], linewidth = 3.5, color = "#5C10E4")
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '', '2020', '', '2021'], alpha = 0.4)
ax3.text(0.816, 2.45, "TRUMP", color = "#5C10E4", fontsize = 15, weight = 'bold', transform = plt.gca().transAxes)
ax3.text(0.790, 2.30, "(2017-2021)", alpha = 0.3, weight = 'bold', transform = plt.gca().transAxes)

# Ax4 For for all three presidents
ax4.plot(bush["Time"], bush["rolling_mean"], color = "#FFEA00")
ax4.plot(obama["Time"], obama["rolling_mean"], color = "#06BFF1")
ax4.plot(trump["Time"], trump["rolling_mean"], color = "#5C10E4")
ax4.set_xticks([])

plt.show()
```
